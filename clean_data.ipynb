```python
# Import Dependencies
import numpy as np
import pandas as pd
```

```python
# Data file
vehicles = "data/vehicles.csv"

# Read in the data
vehicles_data = pd.read_csv(vehicles)
vehicles_data
```

```python
vehicles_data.nunique(axis=0)
```

```python
# Show data summary statistics

vehicles_data.describe().apply(lambda s: s.apply(lambda x: format(x, 'f')))
```

```python
# Remove columns where 40% or more of the values are null

NA_val = vehicles_data.isna().sum()
def na_filter(na, threshold = .4): 
    col_pass = []
    for i in na.keys():
        if na[i]/vehicles_data.shape[0]<threshold:
            col_pass.append(i)
    return col_pass
vehicles_data_cleaned = vehicles_data[na_filter(NA_val)]
vehicles_data_cleaned.columns
```

```python
# IQR

vehicles_data_cleaned = vehicles_data_cleaned[vehicles_data_cleaned['price'].between(999.99, 250000)]
Q1 = vehicles_data_cleaned['price'].quantile(0.25)
Q3 = vehicles_data_cleaned['price'].quantile(0.75)
IQR = Q3 - Q1

# Filtering Values between Q1-1.5IQR and Q3+1.5IQR

vehicles_data_filtered = vehicles_data_cleaned.query('(@Q1 - 1.5 * @IQR) <= price <= (@Q3 + 1.5 * @IQR)')
vehicles_data_filtered.boxplot('price')
```

```python
# Describe semi-cleaned data

vehicles_data_filtered.describe().apply(lambda s: s.apply(lambda x: format(x, 'f')))
```

```python
# year can't be newer than 2020
vehicles_data_filtered = vehicles_data_filtered[vehicles_data_filtered['year'].between(1900, 2020)]

# odometer can't be higher
vehicles_data_filtered = vehicles_data_filtered[vehicles_data_filtered['odometer'].between(0, 271431.5)]
```

```python
# Drop unnecessary columns
vehicles_final = vehicles_data_filtered.copy().drop(['id', 'url', 'region_url', 'image_url', 'region', 'description', 'model', 'state', 'paint_color'], axis=1)


```
